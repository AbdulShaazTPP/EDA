```python
import plotly.express as px

fig = px.histogram(
    d1, 
    y='Screen On Time (hours/day)', 
    x='Operating System', 
    title='OS vs Screen Time'
)

fig.update_layout(
    xaxis_title="Operating System",
    yaxis_title="Screen Time (in Hours)"
)

fig.show()
```

```python
fig = px.histogram(
    d1, 
    y='App Usage Time (min/day)', 
    x='Device Model', 
    title='Device Model'
)

fig.update_layout(
    xaxis_title="Device",
    yaxis_title="Apps Usage Time (in Hours)"
)

fig.show()
```

```python
fig = px.histogram(d1, x='Operating System', y='Battery Drain (mAh/day)', title="Screen Time and Behaviour")

fig.show()
```
