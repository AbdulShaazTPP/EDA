DATA VISUALIZATION

DEMOGRAPHICS

Age Distribution

```python
fig = px.histogram(
    df, 
    x = 'Age',
    nbins = 20,
    title = 'Age Distribution', 
    labels = {'Age':'Age(Years)', 'count': 'Number of Customers'}
)

fig.update_layout(
    xaxis_title = "Age(Years)",
    yaxis_title = "Number of Customers"
)
fig.update_layout(
    bargap = 0.2
)
fig.show()
```

Gender Distributon

```python
px.pie(
    df, 
    names = 'Gender', 
    title = 'Gender Distribution',
    hole = 0.4, 
    labels = {'Gender': 'Gender'}
)
```

Age Group Distribution

```python
bins = [0, 18, 30, 45, 60, 100]  
labels = ['0-18', '19-30', '31-45', '46-60', '60+'] 

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


print(df[['Age', 'Age Group']].head())
```

```python
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '60+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
age_group_counts = df['Age Group'].value_counts().reset_index()
age_group_counts.columns = ['Age Group', 'Number of People']

fig = px.bar(
    age_group_counts,
    x='Age Group',               
    y='Number of People',      
    title='Age Group Distribution',
    labels={'Age Group': 'Age Group', 'Number of People': 'Number of People'},
    text='Number of People' 
)

fig.update_traces(texttemplate='%{text}', textposition='outside')

fig.update_layout(
    xaxis_title="Age Group",
    yaxis_title="Number of People",
    title_x=0.5, 
    template='plotly'
)
fig.show()
```

Purchase Trend by Gender

```python
fig = px.bar(df,
    x = 'Gender',
    y = 'Purchase Amount (USD)',
    title = 'Purchase Trend by Gender',
    labels = {'Gender':'Gender', 'Purchase Amount(USD)':'Total Purchase Amount(USD)'},
    text = 'Purchase Amount (USD)'
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(
    xaxis_title="Gender",
    yaxis_title="Total Purchase Amount (USD)",
    title_x=0.5,
    template='plotly'
)
```

Item and Category

```python
fig = px.bar(df,
    x = 'Item Purchased',
    y = 'Category',
    title = 'Item X Category',
    labels = {'Gender':'Gender', 'Purchase Amount(USD)':'Total Purchase Amount(USD)'},
)
fig.show()
```

Top 10 Popular Items

```python
popular_items = df['Category'].value_counts().reset_index()
popular_items.columns = ['Category', 'Number of Purchases']
fig = px.bar(
    popular_items.head(10),
    x='Category',
    y='Number of Purchases',
    title='Top 10 Popular Items by Number of Purchases',
    labels={'Category': 'Category', 'Number of Purchases': 'Number of Purchases'},
    text='Number of Purchases' 
)
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.update_layout(
    xaxis_title="Category",
    yaxis_title="Number of Purchases",
    title_x=0.5,
    template='plotly',
    xaxis=dict(categoryorder='total descending')
)
fig.show()
```

Purchase Amount by Category

```python
fig_box = px.box(
    df,
    x='Category',                 
    y='Purchase Amount (USD)',    
    title='Purchase Amount per Category: Box Plot',
    labels={'Category': 'Category', 'Purchase Amount (USD)': 'Purchase Amount (USD)'},
    color='Category'               
)
fig_box.update_layout(
    xaxis_title="Category",
    yaxis_title="Purchase Amount (USD)",
    title_x=0.5, 
    template='plotly'
)
fig_box.show()
```

Discounts Utilization

```python
discount_data = df['Discount Applied'].value_counts().reset_index()
discount_data.columns = ['Discount Status', 'Count']
fig = px.pie(
    discount_data,
    names='Discount Status',  
    values='Count',         
    title='Discounts Utilization: Pie Chart',
    hole=0.4,              
    color='Discount Status',
    labels={'Discount Status': 'Discount Applied', 'Count': 'Number of Transactions'}
)
fig.update_traces(textinfo='percent+label')
fig.update_layout(
    title_x=0.5, 
    template='plotly'
)
fig.show()
```

Popular Payment Method

```python
payment_method_data = df['Payment Method'].value_counts().reset_index()
payment_method_data.columns = ['Payment Method', 'Number of Transactions']

fig = px.bar(
    payment_method_data,
    x='Payment Method',
    y='Number of Transactions',
    title='Popular Payment Methods',
    labels={'Payment Method': 'Payment Method', 'Number of Transactions': 'Number of Transactions'},
    text='Number of Transactions' 
)
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.update_layout(
    xaxis_title="Payment Method",
    yaxis_title="Number of Transactions",
    title_x=0.5, 
    template='plotly',
    xaxis=dict(categoryorder='total descending')
)
fig.show()
```

Shipping 

```python
shipping_data = df['Shipping Type'].value_counts().reset_index()
shipping_data.columns = ['Shipping Type', 'Count']

fig = px.pie(
    shipping_data,
    names='Shipping Type', 
    values='Count',    
    title='Shipping Type Distribution',
    hole=0.4,          
    color='Shipping Type',  
    labels={'Shipping Type': 'Shipping Type', 'Count': 'Number of Shipments'}
)
fig.update_traces(textinfo='percent+label') 
fig.update_layout(
    title_x=0.5, 
    template='plotly'
)
fig.show()
```

Relation between Rating and Purchase Amount

```python
fig = px.scatter(
    df,
    x='Review Rating',              
    y='Purchase Amount (USD)',       
    title='Relationship Between Review Ratings and Purchase Amount',
    color='Review Rating',
    size='Purchase Amount (USD)',  
    hover_data=['Review Rating', 'Purchase Amount (USD)']
)
fig.update_layout(
    xaxis_title="Review Ratings",
    yaxis_title="Purchase Amount (USD)",
    title_x=0.5, 
    template='plotly'
)
fig.show()
```

Discounts and Reviews

```python
fig = px.violin(
    df,
    x='Discount Applied',    
    y='Review Rating',        
    title='Effect of Discounts on Review Ratings',
    color='Discount Applied',    
    box=True,                   
    labels={'Discount Applied': 'Discount Applied', 'Review Rating': 'Review Rating'},
    hover_data=['Discount Applied', 'Review Rating']  
)


fig.update_layout(
    xaxis_title="Discount Applied",
    yaxis_title="Review Rating",
    title_x=0.5, 
    template='plotly'
)
fig.show()
```
